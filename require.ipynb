```python
with open('old_requirements.txt', 'r') as file:
    text = file.read()
print(text)
```

```python
# Remove unwanted characters from each line and save to a new file
cleaned_lines = []
for line in text.splitlines():
    # Example: remove digits and '=' characters
    cleaned_line = line.split('==')[0]
    cleaned_lines.append(cleaned_line)

print(cleaned_lines)
```

```python
with open('requirements.txt', 'w') as f:
    f.write('\n'.join(cleaned_lines))
```
